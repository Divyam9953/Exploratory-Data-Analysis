## Top 10 batsmen 

- Based on runs

```python
y = pd.merge(player,ball_by_ball,left_on='Player_Id',right_on='Striker_Id')
```

```python
y.groupby('Player_Name')['Batsman_Scored'].sum().sort_values(ascending=False).head(10)
```

- Based on sixes

```python
a = y.query('Batsman_Scored == 6')
a.groupby('Player_Name')['Batsman_Scored'].count().sort_values(ascending=False).head(10)
```

- Based on fours

```python
a = y.query('Batsman_Scored == 4')
a.groupby('Player_Name')['Batsman_Scored'].count().sort_values(ascending=False).head(10)
```

- Number of 100

```python
x = y.groupby(['Match_Id','Player_Name'])['Batsman_Scored'].agg(np.sum)
y = pd.DataFrame(x)
y.reset_index(inplace=True)
z = y.query('Batsman_Scored >= 100')
z.groupby('Player_Name')['Batsman_Scored'].count().sort_values(ascending=False).head(10)
```

## Top 10 teams

- Based on number of matches won

```python
x = pd.merge(match,team,left_on='Match_Winner_Id',right_on='Team_Id')
x.groupby('Team_Name')['Match_Winner_Id'].count().sort_values(ascending=False).head(10)
```

- Based on winning the toss and winning the match

```python
x = pd.merge(match,team,left_on='Toss_Winner_Id',right_on='Team_Id')
x.groupby('Team_Name')['Match_Winner_Id'].count().sort_values(ascending=False).head(10)
```

- Based on team winning highest number of Man of the match

```python
x = pd.merge(match,team,left_on='Match_Winner_Id',right_on='Team_Id')
x.groupby('Team_Name')['Man_Of_The_Match_Id'].count().sort_values(ascending=False).head(10)
```
